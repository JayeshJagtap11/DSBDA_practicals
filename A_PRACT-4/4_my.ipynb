```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([95,85,80,70,60])
y=np.array([85,95,70,65,70])
```

```python
model=np.polyfit(x,y,1)
model #coefficient of model
```

```python
predict=np.poly1d(model)
predict(y)
```

```python
y_pre=predict(x)
y_pre
```

```python
from sklearn.metrics import r2_score
#for calcualting performance we use the r square method
r2_score(y,y_pre)
```

```python
y_line=model[1]+model[0]*x
plt.plot(x,y_line,c="r")
plt.scatter(x,y_pre)
plt.scatter(x,y,c="r")
```

```python
data=pd.read_csv('data4 - data4.csv')
data
```

```python
data.shape
```

```python
x=data.iloc[:,:1].values
x
```

```python
y=data.iloc[:,1:2].values
y
```

```python
#splitting the dataset into training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
regrestion=LinearRegression()
regrestion.fit(x_train,y_train)
```

```python
y_pred=regrestion.predict(x_test)
y_pred.shape
```

```python
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,regrestion.predict(x_train),color="blue")
plt.title("Area vs Price(Training set)")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regrestion.predict(x_train), color = 'blue')
plt.title('Area vs Price (Test set)')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```
