```python
from scipy import stats
z=np.abs(stats.zscore(data2["math score"]))
print(z)
```

```python
threshold=0.18
```

```python
outlier_point=np.where(z<threshold)
outlier_point
```

```python
rscore=sorted(data2["reading score"])
rscore
```

```python
q1=np.percentile(rscore,25)
q3=np.percentile(rscore,75)
print(q1,q3)
```

```python
IQR=q3-q1
print(IQR)
```

```python
# lwr_bound=IQR+1.5*q1
# upr_bound=IQR-1.5*q3
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound,upr_bound)
```

```python
outlier=[]
for i in  rscore:
    if(i<lwr_bound or i>upr_bound):
        outlier.append(i)
print(outlier)
```

```python
for i in outlier_point:
    data2.drop(i,inplace=True)
data2
```
