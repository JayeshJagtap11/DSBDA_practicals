```python
rscore=sorted(df["reading score"])
rscore
```

```python
q1=np.percentile(rscore,25)
q3=np.percentile(rscore,75)
print(q1,q3)
```

```python
IQR=q3-q1
lw_bound=q1+(IQR*1.5)
up_bound=q3-(IQR*1.5)
print(lw_bound,up_bound)
```

```python
outlierr =[]
for i in rscore:
    if(i<lw_bound or i>up_bound):
        outlierr.append(i)
print(outlierr)
```

```python
for i in outlier:
    df.drop(i,inplace=True)
df
# handling outlier
```
